```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("./Fuel_Consumption_2000-2022.csv")
data #dataframe
```

```python
engine_size = data["ENGINE SIZE"].values
engine_size
```

```python
cylinders = data["CYLINDERS"].values
cylinders
```

```python
fuel_consumption = data["FUEL CONSUMPTION"].values
fuel_consumption
```

```python
emissions = data["EMISSIONS"].values
emissions
```

```python
fig, axs = plt.subplots(nrows=6, ncols= 1,figsize=(10,12))
axs[0].scatter(engine_size, cylinders, c = "k", s=0.1, marker="*")
axs[1].scatter(engine_size, fuel_consumption, c = "b", s=0.1, marker="*")
axs[2].scatter(engine_size, emissions, c = "r", s=0.1, marker="*")
axs[3].scatter(cylinders, fuel_consumption, c = "b", s=0.1, marker="*")
axs[4].scatter(cylinders, emissions, c = "g", s=0.1, marker="*")
axs[5].scatter(fuel_consumption, emissions, c = "r", s=0.1, marker="*")
```
