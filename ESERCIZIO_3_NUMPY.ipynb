#### simulare diecimila partite ciascuna composta da 1000 lanci di una moneta.In quante partite si riscontra almeno il 55% di croci sui 1000 lanci?

```python
numero_partite = 10000
numero_partite
```

## ESERCIZIO 5 PAGINA 1077
#### Considera un array monodimensionale di numeri e trova la posizione di tutti i massimi locali. Per esempio, dato l’array a = [0, 2, 1, 1, 3, 2], 2 e 3 sono massimi locali rispettivamente nelle posizioni 1 e 4. 

```python
array = np.array([0,2,1,1,3,2])
array
```

```python
posizioni_massimi = []
```

```python
array_dx = np.roll(array, +1)
array_dx
```

```python
array_sx = np.roll(array, -1)
array_sx
```

```python
for i in range(1, len(array)-1):
    if[(array[i] > array_sx) & array[i] > array_dx]:
        posizioni_massimi.append(i) 


posizioni_massimi
```

## ESERCIZIO 10 PAGINA 107
##### Nel parco vicino alla scuola è installata una centrali- na per la misurazione del livello di polveri sottili PM10 nell’aria. Talvolta la centralina va in uno stato di errore per cui la misurazione delle PM10 fallisce e il dato cor- rispondente riporta il valore 9999,9 all’interno dell’ar- ray delle misurazioni dati. Crea un array di 1000 campioni che generino dei valori secondo una distribuzione normale con valore medio 19 e varianza 5 e supponi che la centralina sia andata in errore 20 volte. Con l’array così creato: i individua se ci sono due dati errati in finestre di al massimo 5 misurazioni consecutive e segnala che la centralina deve essere riparata; i ripulisci i dati dai valori errati e calcola media, mas- simo e minimo

```python
campioni = np.random.normal(19, np.sqrt(5), 1000)
campioni
```

```python
errori = np.random.choice(1000,20, replace = False)
errori
```

```python
campioni[errori] = 9999.9
campioni
```

```python
finestra_errori_consecutivi = 5
k = 0
for i in campioni:
    if i == 9999.9:
        k +=1
        if k >= finestra_errori_consecutivi:
            print("La centralina deve essere aggiustata")
            break
        else:
            k = 0
```

```python
campioni_puliti = campioni[campioni != 9999.9]
campioni_puliti
```

```python
max = np.max(campioni_puliti)
min = np.min(campioni_puliti)
```

```python
print("massimo:", max)
print("minimo:", min)
```
