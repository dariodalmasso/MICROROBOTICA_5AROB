```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
size_x = 100
size_y = 100
image = np.zeros((size_x, size_y), dtype= np.int16)
```

```python
image[0:5,:] = 255
image[-5:,:] = 255
image[:,0:5] = 255
image[:,-5:] = 255
```

```python
plt.imshow(image)
```

```python
image[size_x//2, size_y //2] = 100
```

```python
plt.imshow(image)
```

```python
size_x = 100
size_y = 100
image = np.zeros((size_x, size_y), dtype= np.int16)
image[:] = np.arange(0,size_y)
image
```

```python
plt.imshow(image, cmap="hsv")
```

```python
size_x = 100
size_y = 100
image = np.zeros((size_x, size_y), dtype= np.int16)
```

```python
image[:] = np.arange(0,size_y)
n_dots = 500
x = np.random.randint(0,size_x, n_dots)
y = np.random.randint(0,size_y, n_dots)
image[x,y] = 255
```

```python
plt.imshow(image, cmap="magma")
```

```python
image[image % 2 == 1] = 0 #numeri dispari a 0
```

```python
plt.imshow(image, cmap="magma")
```
