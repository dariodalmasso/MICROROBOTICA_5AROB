```python
file = pd.read_csv("./world_economic_indicators.csv")
file
```

```python
Paesi = file["Country Name"].unique()
Paesi
```

```python
#2 Calcola il prodotto interno lordo del mondo (ottenuto sommando quello di tutti i paesi) per ogni anno tra il 1960 e il 2022
gdp = file['GDP (current US$)_x'].sum()
gdp
```

```python
#3 Calcola il valore massimo della crescita percentuale del prodotto interno lordo per gli Stati Uniti.

max = file[file["Country Name"] == "United States"]["GDP growth (annual %)_x"].max()
max
```

```python
#4 Estrai la percentuale di disocupazione in Italia nel 2022.
italia = file[(file['Country Name'] == 'Italy') & (file['Year'] == 2022)]["Unemployment, total (% of total labor force)"]
italia
```

```python
#5 Estrai il prodotto interno lordo dell'Italia in tutti gli anni tra il 1960 e il 2022.
gdp_italia = file[file['Country Name'] == 'Italy'][['Year', 'GDP (current US$)_x']]
gdp_italia
```

```python
#6 Con i dati del punto 5) crea un grafico a linea che rappresenti l'andamento nel tempo del prodotto interno lordo dell'Italia. Il grafico deve avere etichette degli assi corrette, titolo e griglia.
fig, ax=plt.subplots(figsize=(8,4))
ax.plot(gdp_italia['Year'], gdp_italia['GDP (current US$)_x'], marker='*', color='black')
ax.set_title('Andamento del Prodotto Interno Lordo Italia (1960-2022)')
ax.set_xlabel('Anno')
ax.set_ylabel('GDP (current US$)')
ax.grid()
plt.show()  
```

```python
gdp_francia = file[file['Country Name'] == 'France'][['Year', 'GDP (current US$)_x']]
```

```python
fig, ax=plt.subplots(figsize=(8,4))
ax.plot(gdp_italia['Year'], gdp_italia['GDP (current US$)_x'], marker='*', color='black')
ax.plot(gdp_francia['Year'], gdp_francia['GDP (current US$)_x'], marker='*', color='red')
ax.set_title('Andamento del Prodotto Interno Lordo Italia e Francia (1960-2022)')
ax.set_xlabel('Anno')
ax.set_ylabel('GDP (current US$)')
ax.grid()
plt.show()  
```
